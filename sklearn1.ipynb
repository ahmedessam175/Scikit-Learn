```python
# 4-confusion matrics
# TP  FP 
# TN  FN
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

y_test=[5,3,9,6,7,5]
y_pred=[4,2,5,4,2,4]

cm=confusion_matrix(y_test , y_pred )
sns.heatmap(cm , center=True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
y_test=[1,1,1,1,0,0,1,0]
y_pred=[1,0,1,1,1,1,0,1]
cm=confusion_matrix(y_test , y_pred )
print(cm)
sns.heatmap(cm , center=True)
plt.show()
```

```python
# 5-accuracy score (TP +TN)/(TP +TN +FP+FN)
from sklearn.metrics import accuracy_score
y_test=[0,2,1,3,5,3]
y_pred=[0,1,2,3,5,3]
acc_sco=accuracy_score(y_test , y_pred , normalize=True)
acc_sco
```

```python
# 6-F1 score 
from sklearn.metrics import f1_score
y_test=[0,2,1,0,0,1]
y_pred=[0,1,2,0,1,2]
f1_s=f1_score(y_test , y_pred ,average='micro')
f1_s
```

```python
# 7-Recall score   (TP)/ Float(TP + FN)
from sklearn.metrics import recall_score
y_test=[0,2,1,0,0,1]
y_pred=[0,1,2,0,1,2]
recall_s=recall_score(y_test , y_pred ,average=None)
recall_s
```

```python
# 7-perscssion score   (TP)/ Float(TP + FP)
from sklearn.metrics import precision_score
y_test=[0,2,1,0,0,1]
y_pred=[0,1,2,0,1,2]
perc_s=precision_score(y_test , y_pred ,average='micro')
perc_s
```

```python
# 9-precision_Recall_fscore_support

from sklearn.metrics import precision_recall_fscore_support 
y_test=['cat','big','dog','cat']
y_pred=['cat','dog','big','cat']
z=precision_recall_fscore_support(y_test , y_pred ,average='micro')
z 
```

```python
# 10-precision_Recall_curve

from sklearn.metrics import precision_recall_curve
y_test=np.array([0,0,1,1])
y_pred=np.array([0.1,0.4,0.35,0.8])
precision,recall,thresholds=precision_recall_curve(y_test , y_pred )
print(precision)
print(recall)
print(thresholds)
```

```python
#11- classification_report

from sklearn.metrics import classification_report
y_test=[0,1,2,2,2,5]
y_pred=[0,0,2,2,1,0]
classif=classification_report(y_test , y_pred )
print(classif)
```

```python
#12- roc_curve
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
y_test=np.array([0,0,1,1])
y_pred=np.array([0.1,0.4,0.35,0.8])
fpr ,tpr,thresholds =precision_recall_curve(y_test , y_pred )
print(precision)
print(recall)
print(thresholds)
```

## <font color="orange">Feature_Selection

```python
#1 SelectPercentile
#اختيار الفيتشرز المطلوبه والمؤثره واستبعاد الباقيين
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectPercentile, chi2
data=load_digits
X, y = load_digits(return_X_y=True)
print(X.shape)
print('*'*15)
X_new = SelectPercentile(chi2, percentile=10).fit_transform(X, y)
print(X_new.shape)
print('*'*15)
```

```python
#2 GenericUnivariateSelect
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import GenericUnivariateSelect, chi2
X, y = load_breast_cancer(return_X_y=True)
print(X.shape)
transformer = GenericUnivariateSelect(chi2, mode='k_best', param=20)  #param=20=20columns
X_new = transformer.fit_transform(X, y)
print(X_new.shape)
print(transformer.get_support())
print(X.shape)
```

```python
# 3-SelectKBest 
from sklearn.datasets import load_digits
from sklearn.feature_selection import SelectKBest, chi2
X, y = load_digits(return_X_y=True)
print(X.shape)

X_new = SelectKBest(chi2, k=10).fit_transform(X, y)
print(X_new.shape)
```

```python
#4- select from model
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import AdaBoostRegressor
from sklearn.datasets import load_boston
from numpy import array

boston = load_boston()
x = boston.data
y = boston.target

print("Feature data dimension: ", x.shape)

estimator = AdaBoostRegressor(random_state=0, n_estimators=50)
selector = SelectFromModel(estimator)
selector = selector.fit(x, y)

# all  true and false
status = selector.get_support()
print("Selection status: ", status)

# all fact feature
features = array(boston.feature_names)
print("All features:")
print(features)

# featue is choosen already
print("Selected features:")
print(features[status])
selector.transform(x)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
X = [[ 0.87, -1.34,  0.31 ],
     [-2.79, -0.02, -0.85 ],
     [-1.34, -0.48, -2.55 ],
     [ 1.92,  1.48,  0.65 ]]
y = [0, 1, 0, 1]

selector = SelectFromModel(estimator=LogisticRegression()).fit(X, y)
print(selector.estimator_.coef_)

print(selector.threshold_)

print(selector.get_support())
print(selector.transform(X))
```

## <font color='orange'>Data Preprocessing

### 1- Data Scaling

```python
# stander scaler  القيمه- المتوسط / الانحراف المعياري
from sklearn.preprocessing import StandardScaler
data = [[0, 0], [0, 0], [1, 1], [1, 1]]
scaler = StandardScaler()
print(scaler.fit(data))
StandardScaler()
print(scaler.mean_)
print('*'*20)
print(scaler.transform(data))
```

### 2- Min Max Scaler

```python
# min max scaler  القيمه- المتوسط / الرينج
from sklearn.preprocessing import MinMaxScaler
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = MinMaxScaler(feature_range=(0,2))
scaler.fit(data)
print(scaler.data_max_)
print('*'*20)
DS=scaler.transform(data)
print(DS)
print('*'*20)
print(scaler.transform([[2, 2]]))
```

### 3- Normalize

```python
from sklearn.preprocessing import normalize
X = [[-2, 1, 2], [-1, 0, 1]]
print(normalize(X, norm="l1"))  # L1 normalization each row independently
print('*'*20)
print(normalize(X, norm="l2"))
```

### 4-Binarizer

```python
from sklearn.preprocessing import Binarizer
B=Binarizer()
X = [[ 1., -1.,  2.],
[ 2.,  0.,  0.],
[ 0.,  1., -1.]]
B.fit(X)  # fit does nothing.
data_bin=B.transform(X)
print(data_bin)
```

### 5-Polynomial

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
X = np.arange(6).reshape(3, 2)
print(X)
print('*'*15)

poly = PolynomialFeatures(2)
print(poly.fit_transform(X))
print('*'*15)
poly = PolynomialFeatures(interaction_only=True)
print(poly.fit_transform(X))
```

## Data Spliting

### 1-Train,Test,Split

```python
import numpy as np
from sklearn.model_selection import train_test_split

X, y = np.arange(10).reshape((5, 2)), range(5)
print(X)
print(list(y))
print('-'*30)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train)
print(y_train)
print(y_test)
```

### 2-KFold

```python
import numpy as np
from sklearn.model_selection import KFold
X = np.array([[1, 2], [3, 4], [1, 2], [3, 4]])
y = np.array([1, 2, 3, 4])
kf = KFold(n_splits=2)
kf.get_n_splits(X)

print(kf)

for i, (train_index, test_index) in enumerate(kf.split(X)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")
```

### 3-leave one out

```python
# leave one out
import numpy as np
from sklearn.model_selection import LeaveOneOut
X = np.array([[1, 2], [3, 4]])
y = np.array([1, 2])
loo = LeaveOneOut()
print(loo.get_n_splits(X))

print(loo)
LeaveOneOut()
for i, (train_index, test_index) in enumerate(loo.split(X)):
    print(f"Fold {i}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")
```
